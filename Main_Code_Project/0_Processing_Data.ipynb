```python
import os
import pandas as pd
import numpy as np
```

```python
# Define the base directory as the current working directory
base_dir = os.getcwd()

# Construct the full path to each data file using relative paths
data_dir = os.path.join(base_dir, '..', 'Data')

# Construct the full path to each data file using relative paths
gaji_bersih_path = os.path.join(base_dir, '..', 'Data', 'Rata-rata Upah_Gaji Bersih Sebulan Buruh_Karyawan_Pegawai Menurut Provinsi dan Lapangan Pekerjaan Utama, 2021.xlsx')
output_per_hour_path = os.path.join(base_dir, '..', 'Data', 'GDP_2HRW_NOC_NB_A-20241224T1208.xlsx')
upah_minimum_regional_path = os.path.join(base_dir, '..', 'Data', 'Upah Minimum Regional_Propinsi, 2019-2020.xlsx')
minimum_wage_global_path = os.path.join(base_dir, '..', 'Data', 'EAR_4MMN_CUR_NB_A-filtered-2024-12-24.xlsx')

# Load the datasets
gaji_bersih_persektor = pd.read_excel(gaji_bersih_path)
output_per_hour = pd.read_excel(output_per_hour_path)
upah_minimum_regional = pd.read_excel(upah_minimum_regional_path)
minimum_wage_global = pd.read_excel(minimum_wage_global_path)
```

```python
gaji_bersih_persektor
```

```python
# Rename columns to more descriptive names
gaji_bersih_persektor.rename(columns={
    'A': 'Pertanian_Kehutanan_Perikanan',
    'B': 'Pertambangan_Penggalian',
    'C': 'Industri_Pengolahan',
    'D': 'Pengadaan_Listrik_Gas',
    'E': 'Pengadaan_Air_Pengelolaan_Sampah_Limbah_Daur_Ulang',
    'F': 'Konstruksi',
    'G': 'Perdagangan_Besar_Eceran_Reparasi_Perawatan_Mobil_Motor',
    'H': 'Transportasi_Pergudangan',
    'I': 'Penyediaan_Akomodasi_Makan_Minum',
    'J': 'Informasi_Komunikasi',
    'K': 'Jasa_Keuangan_Asuransi',
    'L': 'Real_Estat',
    'M,N': 'Jasa_Perusahaan',
    'O': 'Administrasi_Pemerintahan_Pertahanan_Jaminan_Sosial_Wajib',
    'P': 'Jasa_Pendidikan',
    'Q': 'Jasa_Kesehatan_Kegiatan_Sosial',
    'R,S,T,U': 'Jasa_Lainnya'
}, inplace=True)


gaji_bersih_persektor
```

```python
type('Tahun')
```

```python
# Convert 'Tahun' column to datetime format
gaji_bersih_persektor['Tahun'] = pd.to_datetime(gaji_bersih_persektor['Tahun'], format='%m-%Y')
# Create a new column for months only
gaji_bersih_persektor['Month'] = gaji_bersih_persektor['Tahun'].dt.month
gaji_bersih_persektor
```

```python
# Check for missing values
print(gaji_bersih_persektor.isnull().sum())
print(output_per_hour.isnull().sum())
print(upah_minimum_regional.isnull().sum())
print(minimum_wage_global.isnull().sum())
```

```python
# Fill missing values with mean for numerical columns
gaji_bersih_persektor['Pengadaan_Listrik_Gas'].fillna(gaji_bersih_persektor['Pengadaan_Listrik_Gas'].mean(), inplace=True)
gaji_bersih_persektor['Pengadaan_Air_Pengelolaan_Sampah_Limbah_Daur_Ulang'].fillna(gaji_bersih_persektor['Pengadaan_Air_Pengelolaan_Sampah_Limbah_Daur_Ulang'].mean(), inplace=True)
gaji_bersih_persektor['Informasi_Komunikasi'].fillna(gaji_bersih_persektor['Informasi_Komunikasi'].mean(), inplace=True)
gaji_bersih_persektor['Real_Estat'].fillna(gaji_bersih_persektor['Real_Estat'].mean(), inplace=True)
```
